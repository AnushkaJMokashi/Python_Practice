```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7,]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,'--r*')
```

```python
plt.plot(x,y,color = 'red',marker='D', linestyle=' ')  #hxadecimal is also allowed
```

```python
plt.plot(x,y,color = 'red',marker='+', linestyle=' ',markersize=20)  
```

```python
#alpha is for opacity
plt.plot(x,y,color = 'red',marker='+', linestyle=' ',markersize=20,alpha = 0.5)  
```
