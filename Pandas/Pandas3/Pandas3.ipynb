```python
df = pd.read_excel("weather_data.xlsx","Sheet1")
df
```

```python
weather_data = {
    'day': ['1/1/2012','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature': [32,35,28,24,32,31],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain','Sunny']
}
df2 = pd.DataFrame(weather_data)
df2
```

```python
#using tuples list 
weather_data = [
    ('1/1/2012',32,6,'Rain'),
    ('1/2/2012',30,5,'Sunny'),
    ('1/3/2012',20,6,'Snow')
]

df = pd.DataFrame(weather_data,columns= ["day","temp","ws","event"])
df
```

```python
#using list of dictionaries
weather_data = [
    {'day':'1/1/2012','temp':32,'ws':6,'event':'Rain'},
    {'day':'1/2/2012','temp':30,'ws':7,'event':'Sunny'},
    {'day':'1/3/2012','temp':22,'ws':6,'event':'Snow'}
]

df = pd.DataFrame(weather_data)
df
```

```python
df2 = df.to_csv('weather_data.csv', index = True)
```
