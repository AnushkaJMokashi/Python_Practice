```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather_data_pandas6.csv")
df = df[0:7]
df
```

```python
#replace two different values
new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
df
```

```python
#replace based on values
new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : -99999,
    'event' : '0'
},np.NaN)
new_df
```

```python
new_df = df.replace({
    -99999: np.NaN,
    '0' : 'Sunny'
})
new_df
```
