#  Handling Missing Data

__Topics:__
* fillna to fill missing values using different ways
* interpolate to make a guess on missing values using interpolation
* dropna to drop rows with missing values

```python
#df.set_index('day',inplace=True)
df.reset_index('day',inplace=True)
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
#replace NaN values
```

```python
new_df=df.fillna(0)
```

```python
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': "no event"
})
```

```python
new_df
```

```python
#loking at temperature it is unrealistic to drop temp that much hence we can
#carry forward the previous value for NaN value
new_df = df.fillna(method = "ffill")
new_df
```

```python
#backward fill
new_df = df.fillna(method = "bfill")
new_df
```

```python
#copy along columns
new_df = df.fillna(method = "bfill",axis="columns")
new_df
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
#carry forward only once
new_df = df.fillna(method = "ffill",limit=1)
new_df
```

```python
#interpolate (linear)
new_df = df.interpolate(method = "time")
new_df
#observing temperature columns we can see that now the values are reasonable
```

```python
#drop all rows with nan values
new_df = df.dropna()
new_df
```

```python
#drop only with criteria
new_df = df.dropna(how="all")
new_df
```

```python
#only 1 na then keep the row
new_df = df.dropna(thresh =1)
new_df
```

```python
new_df = df.dropna(thresh =2)
new_df
```
